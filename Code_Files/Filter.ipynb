```
def load_images(image_path): # load all images of the input folder in list icludes pathes of these images
  image_files = sorted([os.path.join(image_path, file)
                          for file in os.listdir(image_path)])
  return image_files
```

```
def copy_files(path_txt, images, dest_path):
    for i in range(len(images)):
      path = path_txt + '/' +Path(images[i]).stem + '.txt'
      if (os.path.isfile(path)):
        shutil.copy(path, dest_path)
      else:
        print(path)
```
